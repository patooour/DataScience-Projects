### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user',sep='|',index_col='user_id')
users
```

### Step 4. Discover what is the mean age per occupation

```python
groupUser = users.groupby('occupation')
groupUser.mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
a=users.groupby(['occupation','gender']).gender.count()
b=users.groupby('occupation').gender.count() #revision on this 
c=(a/b)*100
c.sort_values(ascending=False).xs('M',level=1)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
groupUser[['age']].agg(['min','max'])
```
